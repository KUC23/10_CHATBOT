bbc 크롤링

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup

def scrape_bbc_news_with_selenium(category_url):
    # Selenium WebDriver 설정
    options = Options()
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

    driver.get(category_url)
    
    # 특정 요소가 로드될 때까지 대기
    try:
        WebDriverWait(driver, 20).until(EC.presence_of_all_elements_located((By.CLASS_NAME, '//*[@id="main-content"]/article/section[1]/div/div[2]/div[1]/div/div/div[1]/a/div/div[2]/div[1]/div/h2')))
    except Exception as e:
        print(f"Error: {e}")
        driver.quit()
        return []

    # 페이지 HTML을 BeautifulSoup로 파싱
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    articles = []

    # 기사 추출 
    for article in soup.find_all('a', class_='//*[@id="main-content"]/article/section[1]/div/div[2]/div[1]/div/div/div[1]/a/div/div[2]/div[1]/div/h2'):
        title = article.get_text()  # 기사 제목
        link = article['href']  # 기사 링크

        # 링크가 상대 경로일 경우 절대 경로로 변환
        if not link.startswith("http"):
            link = f"https://www.bbc.com{link}"
        articles.append({"title": title, "link": link})

    driver.quit()  # 브라우저 종료
    return articles

# 카테고리별로 크롤링하기
categories = {
    "main": "https://www.bbc.com/news/world",
    "technology": "https://www.bbc.com/news/technology",
    "business": "https://www.bbc.com/news/business",
    "science": "https://www.bbc.com/news/science_and_environment",
    "health": "https://www.bbc.com/news/health",
    "politics": "https://www.bbc.com/news/politics",
    "art": "https://www.bbc.com/news/entertainment_and_arts",
    "sport": "https://www.bbc.com/sport"
}

# 각 카테고리에서 기사를 크롤링하고 출력하기
for category, url in categories.items():
    print(f"Fetching articles for category: {category}")
    articles = scrape_bbc_news_with_selenium(url)
    
    if articles:  # 기사가 있을 경우 출력
        for article in articles[:5]:  # 상위 5개 기사만 출력
            print(f"Title: {article['title']}\nLink: {article['link']}\n")
    else:
        print(f"No articles found for category: {category}")
    print("="*50)
```
