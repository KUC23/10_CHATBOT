```python
import requests
import pandas as pd
import time

API_KEY = 'wG9K6FxHHSExgFSNb97LKtjAAOX7umc3'

# 뉴욕타임스 API 카테고리 매핑
categories = {
    "world": "world",
    "technology": "technology",
    "business": "business",
    "science": "science",
    "health": "health",
    "politics": "politics",
    "entertainment": "arts",
    "sport": "sports"
}

def fetch_nyt_news(category):
    """특정 카테고리의 뉴욕타임스 기사를 가져오는 함수."""
    section = categories.get(category, "home")
    url = f"https://api.nytimes.com/svc/topstories/v2/{section}.json"
    params = {'api-key': API_KEY}
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        articles = data.get('results', [])
        print(f"\nFetching articles for category: {category}")
        
        # 기사 데이터 필터링 및 변환
        filtered_articles = []
        for article in articles[:5]:  # 각 카테고리 상위 5개 기사만 처리
            filtered_articles.append({
                'Category': category,
                'Title': article.get('title', 'No Title'),
                'Abstract': article.get('abstract', 'No Abstract'),
                'URL': article.get('url', 'No URL'),
                'Published Date': article.get('published_date', 'No Date'),
                'Section': article.get('section', 'No Section')
            })
        
        return filtered_articles
    elif response.status_code == 429:
        print(f"Too many requests for category: {category}. Please wait.")
    else:
        print(f"Failed to fetch articles for category: {category}, Status Code: {response.status_code}")
    
    return []  # 요청 실패 시 빈 리스트 반환

# 전체 뉴스 데이터를 저장할 리스트
all_news = []

# 각 카테고리에 대해 뉴스 크롤링
for category in categories.keys():
    category_news = fetch_nyt_news(category)
    all_news.extend(category_news)
    time.sleep(2)  # 요청 간 2초 대기

# 데이터프레임으로 변환 및 CSV 저장
if all_news:
    df = pd.DataFrame(all_news)
    
    # CSV 파일로 저장
    csv_file_name = 'nyt_news.csv'
    df.to_csv(csv_file_name, index=False, encoding='utf-8-sig')
    print(f"\nSaved {len(all_news)} articles to {csv_file_name}")
else:
    print("No news data to save.")
```
